```python
import requests

# Assuming 'movie_id' column holds the movie IDs
movie_ids = df['id']

# Defining new columns for the top 3 actors
df['actor_1'] = ""
df['actor_2'] = ""
df['actor_3'] = ""

for movie_id in movie_ids:
    url = f"https://api.themoviedb.org/3/movie/{movie_id}/credits?language=en-US"

    headers = {
        "accept": "application/json",
        "Authorization": "Bearer ezIK5H7Xbvc6Hc"
    }

    response = requests.get(url, headers=headers)
    credits_data = response.json()

    # Extracting top 3 actors' names
    actors = []
    if 'cast' in credits_data:
        actors = [actor['name'] for actor in credits_data['cast'][:3]]  # Considering first 3 cast members

    # Storing top 3 actors' names in the DataFrame
    df.at[movie_id, 'actor_1'] = actors[0] if len(actors) > 0 else ""
    df.at[movie_id, 'actor_2'] = actors[1] if len(actors) > 1 else ""
    df.at[movie_id, 'actor_3'] = actors[2] if len(actors) > 2 else ""

df
```
