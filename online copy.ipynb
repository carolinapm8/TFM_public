```python
#combino los géneros en una lista para pasarlos a variables numéricas
import pandas as pd
columnas_gen = ['columna1', 'columna2', 'columna3', 'columna4', 'columna5', 'columna6', 'columna7']
dfg['all_genres'] = dfg[columnas_gen].apply(lambda x: x.dropna().tolist(), axis=1)
```

```python
#codifico los géneros
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
generos_encoded = mlb.fit_transform(dfg['all_genres'])

# Crear un DataFrame con las columnas codificadas
genres_df = pd.DataFrame(generos_encoded, columns=mlb.classes_, index=dfg.index)

# Concatenar las nuevas columnas codificadas al DataFrame original
dfg = pd.concat([dfg, genres_df], axis=1)

# Eliminar las columnas originales de géneros
dfg.drop(columns=columnas_gen + ['all_genres'], inplace=True)
```

```python
#veo la popularidad y el vote_average por género 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

genres = ['Action', 'Adventure', 'Animation', 'Comedy', 'Crime', 'Documentary', 'Drama', 
          'Family', 'Fantasy', 'History', 'Horror', 'Music', 'Mystery', 'Romance', 
          'Science Fiction', 'TV Movie', 'Thriller', 'War', 'Western']

genre_data = pd.DataFrame()

for genre in genres:
    genre_df = dfg[dfg[genre] == 1][['vote_average', 'popularity']]
    genre_df['genre'] = genre
    genre_data = pd.concat([genre_data, genre_df])

plt.figure(figsize=(14, 8))

sns.scatterplot(data=genre_data, x='vote_average', y='popularity', hue='genre', palette='tab20', alpha=0.7)

plt.title('Vote Average vs Popularity x género')
plt.xlabel('Vote Average')
plt.ylabel('Popularity')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
```

```python
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline

# Identificar las columnas numéricas y categóricas (excluyendo las ya codificadas)
num_features = dfg.select_dtypes(include=['int64', 'float64']).columns.tolist()
cat_features = dfg.select_dtypes(include=['object']).columns.tolist()

# Eliminar la columna objetivo 'budget' de las características numéricas
num_features.remove('budget')

# Manejar valores faltantes y codificar variables categóricas
num_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median'))
])

cat_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_pipeline, num_features),
        ('cat', cat_pipeline, cat_features)
    ], remainder='passthrough')  # Mantener las otras columnas sin cambios
```
