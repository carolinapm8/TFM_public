```python
import pandas as pd

#paso los decimales a enteros
columnas = ['columna1', 'columna2', 'columna3', 'columna4', 'columna5', 'columna6', 'columna7']
for col in columnas:
    df2[col] = pd.to_numeric(df2[col], errors='coerce')  # 'coerce' convierte los valores no numéricos a NaN

#le pongo un 0 a lo que no tiene numero 
df2[columnas] = df2[columnas].fillna(0).astype(int)
df2
```

```python
import pandas as pd

#diccionario para mapear las columnas con el num al que pertence cada genero
mapeo_generos = {
    0: 'No_gender',
    28: 'Action', 12: 'Adventure', 16: 'Animation', 35: 'Comedy', 80: 'Crime',
    99: 'Documentary', 18: 'Drama', 10751: 'Family', 14: 'Fantasy', 36: 'History',
    27: 'Horror', 10402: 'Music', 9648: 'Mystery', 10749: 'Romance', 878: 'Science Fiction',
    10770: 'TV Movie', 53: 'Thriller', 10752: 'War', 37: 'Western'
}

columnas = ['columna1', 'columna2', 'columna3', 'columna4', 'columna5', 'columna6', 'columna7']
df2[columnas] = df2[columnas].replace(mapeo_generos)

```
